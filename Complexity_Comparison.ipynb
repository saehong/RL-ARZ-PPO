```python
import scipy.io as sio
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from settings_file import *
```

```python
M_40_data = sio.loadmat('ARZ_python_result_RL_Outlet_M_40.mat')
M_40_simtime = 2910.3300065994263
M_40_dt = 0.25/50*40

M_50_data = sio.loadmat('ARZ_python_result_RL_Outlet_M_50.mat')
M_50_simtime = 3180.550197124481
M_50_dt = 0.25

M_60_data = sio.loadmat('ARZ_python_result_RL_Outlet_M_60.mat')
M_60_simtime = 3455.649220228195
M_60_dt = 0.25/50*60

T=settings['T']

M_40_time_vec = np.arange(0,T,M_40_dt)
M_50_time_vec = np.arange(0,T,M_50_dt)
M_60_time_vec = np.arange(0,T,M_60_dt)
```

```python

fig, ax1 = plt.subplots(figsize=(8,6))

plt.plot(M_40_time_vec,M_40_data['rwd_RL'][0],linewidth=4,label='M=40')
plt.plot(M_50_time_vec,M_50_data['rwd_RL'][0],linewidth=4,label='M=50')
plt.plot(M_60_time_vec,M_60_data['rwd_RL'][0][:-1],linewidth=4,label='M=60')
plt.xlabel('Episode Time [sec]',Fontsize=16)
plt.ylabel('Reward',Fontsize=16)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.grid(True)
plt.legend(fontsize=16)
plt.show()

fig.savefig("M_Reward_Comparison.pdf", bbox_inches='tight')
```

```python
sim_time = []
sim_time.append(M_40_simtime)
sim_time.append(M_50_simtime)
sim_time.append(M_60_simtime)



fig, ax1 = plt.subplots(figsize=(8,6))
name = ['M=40','M=50','M=60']
plt.bar(name,sim_time)
plt.xlabel('Cases',Fontsize=16)
plt.ylabel('Computation Time [Sec]',Fontsize=16)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.show()

fig.savefig("M_Time_Comparison.pdf", bbox_inches='tight')
```
